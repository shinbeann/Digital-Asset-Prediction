```python
import pandas as pd
import os
```

```python
# Get all .csv files in the current directory
csv_files = [f for f in os.listdir() if f.endswith('_1d.csv')]

# Initialize an empty list to hold dataframes
dataframes = []

# Loop through the csv files and read them into dataframes
for file in csv_files:
    df = pd.read_csv(file)
    # Add a new column with the file name without the suffix
    df['coin_label'] = file.replace('_1d.csv', '')
    dataframes.append(df)

# Concatenate all dataframes into a single dataframe
collated_df = pd.concat(dataframes, ignore_index=True)
```
