```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('/Users/aishwaryaiyer/Documents/GitHub/Digital-Asset-Prediction/data/processed/combined_dataset_v1.csv')
print(df)
```

```python
numerical_columns = df.select_dtypes(include=['float64']).columns

correlation_matrix = df[numerical_columns].corr()

for target_column in numerical_columns:
    print(f"\nCorrelations with {target_column}:")
    print(correlation_matrix[target_column]
          .sort_values(ascending=False)  
          .drop(target_column))  
```
